```python
%pylab inline
import pybedtools
import pandas as pd
from pyfaidx import Fasta
import json

cds = pybedtools.BedTool('../data/D_rerio_z10/annotation/Danio_rerio.GRCz10.90.gffutils.cds.bed').remove_invalid().saveas('Danio_rerio.cds.bed')
cds = pybedtools.BedTool('Danio_rerio.cds.bed').sort().to_dataframe()#.set_index('name')
gene = pybedtools.BedTool('../data/D_rerio_z10/annotation/Danio_rerio.GRCz10.90.gffutils.genes.bed').to_dataframe()#.set_index('name')
utr5 = pybedtools.BedTool('../data/D_rerio_z10/annotation/Danio_rerio.GRCz10.90.gffutils.UTR5.bed').to_dataframe()#.set_index('name')
utr3 = pybedtools.BedTool('../data/D_rerio_z10/annotation/Danio_rerio.GRCz10.90.gffutils.UTR3.bed').to_dataframe()#.set_index('name')
fasta = Fasta('../data/D_rerio_z10/fasta/Danio_rerio.GRCz10.dna.toplevel.fa')

def get_fasta_sequence(fasta, intervals):
    if isinstance(fasta, str):
        fasta = Fasta(fasta)
    sequence = []
    for interval in intervals:
        chrom, start, stop, strand = interval
        if int(start) == 46308516:
            print 'xxxxx'
        if int(stop)-int(start)==1:
            continue
        if strand == '+':
            seq = fasta[chrom][int(start):int(stop)].seq
        elif strand == '-':
            try:
                seq = fasta[chrom][int(start):int(stop)].complement.seq
            except:
                print 'xxx', start, stop
        sequence.append(seq)
    if strand == '-':
        sequence.reverse()# = sequence[::-1]
    return ('').join(sequence).upper()
```
