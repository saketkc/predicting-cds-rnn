```python
%pylab inline
import pandas as pd
import seaborn as sns
sns.set(font_scale=2)
sns.set_style('whitegrid')
```

```python
def train_curve(table, model):
    fig = plt.figure(figsize=(10, 7)) 
    train_df = pd.read_table(table, sep=' ', names=['epoch', 'train_acc'])
    train_df['train_acc'].plot()
    plt.title('Training accuracy ({})'.format(model))
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    
def test_curve(table, model):
    fig = plt.figure(figsize=(10, 7)) 
    train_df = pd.read_table(table, sep=' ', names=['epoch', 'test_acc'])
    train_df['test_acc'].plot()
    plt.title('Testing accuracy ({})'.format(model))
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    
```

# LSTM Embeddding


```python
# LSTM Embeddding

! cat ../notebooks_embedding/train_acc_19000.log| grep Train |  cut -f 2,6 -d ' ' > embedding_lstm_train_acc_19000.tsv
```

```python
! cat ../notebooks_embedding/test_acc_19000.log | cut -f 2,6 -d ' ' > embedding_lstm_test_acc_19000.tsv
```

```python
train_curve('embedding_lstm_train_acc_19000.tsv', 'Embedding + LSTM')
plt.savefig('embedding_lstm_train_acc_19000.pdf', dpi=900)
```

```python
test_curve('embedding_lstm_test_acc_19000.tsv', 'Embedding + LSTM')
plt.savefig('embedding_lstm_test_acc_19000.pdf', dpi=900)
```
